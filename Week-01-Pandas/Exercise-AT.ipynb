## Exercise 2 - Filtering

Return the following subsets of the dataframe.

1. How many listings are there with a price less than 100? 


2. Find how many listings there are in just Brooklyn.


3. Find how many listings there are in Brooklyn with a price less than 100.


4. Using `.isin()` select anyone that has the host name of Michael, David, John, and Daniel.


5. Create a new column called `adjusted_price` that has $100 added to every listing in Williamsburg.  The prices for all other listings should be the same as the were before. 


6. What % of the rooms are private, and what % of the rooms are shared.  
    * Hint, use `.value_counts()`


```python
# 1. How many listings are there with a price less than 100? 22778

# syntax: df.loc['row', 'col']

# df[df['price'] < 100]

price_bel100 = df['price'] < 100
df_price_bel100 = df.loc[price_bel100]

print(f"Listings with price less than 100: {len(df_price_bel100)}")
display(df_price_bel100)
```

```python
# 2. Make a new DataFrame of listings in Brooklyn named `df_bk`
# and find how many listings in just Brooklyn. 18632

# df_bk = df.loc[ : , 'neighbourhood_group'].value_counts()['Brooklyn']
filter_brooklyn = df['neighbourhood_group'] == 'Brooklyn'
df_bk = df.loc[filter_brooklyn]

display(df.loc[ : , 'neighbourhood_group'].value_counts())

print(f"'Brooklyn' in column `neighbourhood_group`: {len(df_bk)}")
display(df_bk)
# df_bk.head()
```

```python
# 3. Find how many listings there are in Brooklyn with a price less than 100. 10473
df_bk_bel100 = df.loc[filter_brooklyn & price_bel100]

print(f"'Brooklyn' in col neighbourhood_group and below 100 in col price: {len(df_bk_bel100)}")
display(df_bk_bel100)
```

```python
# 4. Using `.isin()` select anyone that has the host name of Michael, David, John, and Daniel.
# How many total are there that have those names

# Create filter selection
target_hosts = ['Michael', 'David', 'John', 'Daniel']

# Filter dataframe by selection
filter_hosts = df['host_name'].isin(target_hosts)

# Create new dataframe to reflect filter
df_host4 = df[filter_hosts]['host_name']

display(df_host4.value_counts())
df.loc[filter_hosts]
```

```python
# 5. Create a new column called `adjusted_price` that has $100 added to every listing in Williamsburg. 3486 Williamsburg
# The prices for all other listings should be the same as the were before.

# !!!!!!!!!!!!! ONLY COLUMN LABELS SHOWING MEANS THERE'S NO RESULTS !!!!!!!!!

# Create filter selection
filter_wb = df['neighbourhood'] == 'Williamsburg'

# Filter dataframe by selection
df_wb = df.loc[filter_wb]

# # Sanity check
display(df.loc[ : , 'neighbourhood'].value_counts())

# display(df_wb)

df['adjusted_price'] = np.where(filter_wb, (df['price'] + 100), df['price'])

display(df)
display(df[filter_wb])
```

```python
# 6. What % of the rooms are private, and what % of the rooms are shared.

# Quantity check
display(df.loc[ : , 'room_type'].value_counts())

# Double check length of dataframe
print(sum(df.loc[ : , 'room_type'].value_counts()))
print(len(df))

# Filter 'Private room' from col 'room_type'
filter_priv_room = df.loc[ : , 'room_type'] == 'Private room'
filter_not_priv_room = df.loc[ : , 'room_type'] != 'Private room'

sum_of_df = len(df)
priv_rm_ttl = len(df[filter_priv_room])
nonpriv_rm_ttl = len(df[filter_not_priv_room])

# Check length of priv and non-priv rooms
print(priv_rm_ttl)
print(nonpriv_rm_ttl)

# percent = (target / total) * 100
# format to 2 decimal places
priv_rm_percent = (priv_rm_ttl / sum_of_df) * 100
nonpriv_rm_percent = (nonpriv_rm_ttl / sum_of_df) * 100

print(f"Percent of 'Private room' in col 'room_type': {priv_rm_percent :.2f}")
print(f"Percent of non-'Private room' in col 'room_type': {nonpriv_rm_percent :.2f}")

# -----------------------------------------
# or use value_counts() normalize
print("\n------------------------\n\nUsing `value_counts()`")

display(df.loc[ : , 'room_type'].value_counts())
display(df.loc[ : , 'room_type'].value_counts(normalize = True))
df.loc[ : , 'room_type'].value_counts(normalize = True)['Private room']
```

# Exercise 3 - Grouping

1. Using `groupby`, count how many listings are in each neighbourhood_group.


2. Using `groupby`, find the mean price for each of the neighbourhood_groups. 


3. Using `groupby` and `.agg()`, find the min and max price for each of the neighbourhood_groups. 


4. Using `groupby`, find the median price for each room type in each neighbourhood_group.


5. Using `groupby` and `.agg()`, find the count, min, max, mean, median, and std of the prices for each room type in each neighbourhood_group.

```python
# 1. Using `groupby`, count how many listings are in each neighbourhood_group.

gb_neighborhood_group = df.groupby('neighbourhood_group')
gb_neighborhood_group['neighbourhood_group'].count()
```

```python
# 2. Using `groupby`, find the mean price for each of the neighbourhood_groups.

gb_neighborhood_group['price'].mean()
```

```python
# 2.5. Using `groupby`, find the mean price for each room_type.

gb_room_type = df.groupby('room_type')
gb_room_type['price'].mean()
```

```python
# 3. Using `groupby` and `.agg()`, find the min and max price for each of the neighbourhood_groups.

gb_price_ngroups = df.groupby('neighbourhood_group')
gb_price_ngroups['price'].agg(['min', 'max'])
```

```python
# 4. Using `groupby`, find the mean price for each room_type in each neighbourhood_group.

cols_rmtype_ngroup = ['neighbourhood_group', 'room_type']

gb_rmtype_ngroup = df.groupby(cols_rmtype_ngroup)

gb_rmtype_ngroup['price'].mean()
```

```python
# 5. Using `groupby` and `.agg()`, find the count, min, max, mean, median, and std of the prices
# for each room type in each neighbourhood_group.

gb_rmtype_ngroup['price'].agg(['count', 'min', 'max', 'mean', 'median', 'std'])
```

```python
# 1. Load the `prices.csv` and the `n_listings.csv`

df_prices = pd.read_csv("~/csc-assign/ds24-files/Week-01-Pandas/data/prices.csv")
display(df_prices)

df_listings = pd.read_csv("~/csc-assign/ds24-files/Week-01-Pandas/data/n_listings.csv")
display(df_listings)
```

```python
# 2. Do join that keeps all the records for each table.

df_listings = pd.read_csv("~/csc-assign/ds24-files/Week-01-Pandas/data/n_listings.csv", sep = ';')
display(df_listings)

df_price_listings_outer = pd.merge(df_prices, df_listings, how = 'outer') # full outer join, keep all rows
display(df_price_listings_outer)

df_price_listings_inner = pd.merge(df_prices, df_listings, how = 'inner') # intersection of keys from both frames, merge all common column name
display(df_price_listings_inner)

df_price_listings_cross = pd.merge(df_prices, df_listings, how = 'cross') # cartesian product
display(df_price_listings_cross)
```

```python
# 3. Save your joined csv as `joined.csv` into the data folder.

import os

abs_path = "~/csc-assign/ds24-files/Week-01-Pandas/data/"

merge_price_listings_fname = 'joined.csv'

df_price_listings_outer.to_csv(os.path.join(abs_path, merge_price_listings_fname), index = False)

```

```python
# 4. Load your newly saved file, see if it looks the same.  If not, try saving with argument `index=False`

df_joined = pd.read_csv(os.path.join(abs_path, 'joined.csv'))

display(df_joined)
```
